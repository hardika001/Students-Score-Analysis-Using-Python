```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students_record.csv")
```

 ### Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df.head()
```

### Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analyzed that:
The number of females in the data is more than number of males

### Impact of Parents education on student's score

```python
gb = df.groupby("ParentEduc").agg({"MathScore" : "mean","ReadingScore" : "mean", "WritingScore" : "mean" })
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart we have concluded:
The education of the parents have a good impact on their childrens score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" : "mean","ReadingScore" : "mean", "WritingScore" : "mean" })
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

From the above chart we have analyzed that:
Marital Status of the parents have no or negligible impact on their children's Score

### Student's Score on the basis of Gender

```python
gby = df.groupby("Gender").agg({"MathScore" : "mean","ReadingScore" : "mean", "WritingScore" : "mean"})
print(gby)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gby)
plt.title("Female vs Male")
plt.show()
```

From the above Chart we have analyzed that :
Other than Maths,Female have scored more marks than males

### Impact of Sports on Student's Score

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore" : "mean","ReadingScore" : "mean", "WritingScore" : "mean" })
print(gb2)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb2, annot = True)
plt.title("Impact of Sports on Student's Score")
plt.show()
```

From the above chart we have analyzed that:
Sports have very low impact on Students Score

### Impact of Study Hours on Scores

```python
gb3 = df.groupby("WklyStudyHours").agg({"MathScore" : "mean","ReadingScore" : "mean", "WritingScore" : "mean" })
print(gb3)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb3, annot = True)
plt.title("Impact of Study Hours on Scores")
plt.show()
```

From the above chart we should conclude that:
The Student who study more than 10 hours get more marks

### Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

From the Above Chart we have analyzed that:
There are highest students from Ethnic Group C followed by Group D and Group B
